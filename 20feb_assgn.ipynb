```python
Q1. Explain GET and POST methods.
Ans. GET Method:
    -> It is used to request data from a specific webpage .
    -> It works under HTTP Protocol.
    
    POST Metgod:
    -> It is also used to request adata from a webpage used to be processed by a function.
    -> It also works under HTTP Protocol.
```

```python
Q2. Why is request used in Flask?
Ans. The request object is an instance of Request class used to access the incoming client 
requests in the form of data, headers,cookies etc.
It is used to fetch data from HTML components and use it as per definition in function.
```

```python
Q3. Why is redirect() used in Flask?
Ans. The redirect() function is used to redirect a user to a different page, hyperlink from a 
webpage.
```

```python
Q4. What are templates in Flask? Why is the render_template() function used?
Ans. templates are part of Flask which provides Dynamic HTML effect where the data on webpage can
vary as a template consists of variables , loops .

render_template function is used to send the data to variables defined in HTML file and render
the same with the values passed.
```
